```python
!pip install tabula-py

import tabula
import pandas as pd

# Path to the PDF file
pdf_path = "/content/comm-q2-2024.pdf"

# Extract tables from the PDF on page 5
tables = tabula.read_pdf(pdf_path, pages=5, multiple_tables=True)

# Convert the first table in the list to a DataFrame and limit to the first 27 columns
df = pd.DataFrame(tables[0])
df = df.iloc[:21]

# Display the DataFrame
df

```

```python
#drop first four rows from the dataset
df.drop(index=[0, 1, 2, 3], inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
#Rename the columns
df.columns
```

```python
df.rename(columns = {
    '(millions of Canadian dollars)': "Operating Results",
    'Unnamed: 0': "Personal and Commercial",
    'Unnamed: 1': "Wealth Management",
    'Unnamed: 2': "Financial Markets",
    'Unnamed: 3': "USSF&J",
    'Six months ended April 30': "Total",
    }, inplace = True)
df
```

```python
#drop all the Nan and keep it blank
df.fillna('', inplace=True)
df
```

```python
#drop all the "-" and keep it blank
df.replace('-', '', inplace=True)
df
```

```python
#drop values in '()'
df.replace(r'\([^)]*\)', '', regex=True, inplace=True)
df
```

```python
#preprocessing steps
df.replace(' ', '', inplace=True)
df
```

```python
import pandas as pd


def get_value(row_name, column_name):
  """
  Returns the value at the intersection of the given row and column names.

  Args:
    row_name: The name of the row (e.g., 'Net interest income').
    column_name: The name of the column (e.g., 'Personal and Commercial').

  Returns:
    The value at the intersection, or None if not found.
  """
  try:
    value = df.loc[df['Operating Results'] == row_name, column_name].values[0]
    return value
  except IndexError:
    return None

# Chatbot interaction loop
while True:
  row_name = input("Enter the row name (or 'quit' to exit): ")
  if row_name.lower() == 'quit':
    break

  column_name = input("Enter the column name: ")

  value = get_value(row_name, column_name)
  if value is not None:
    print(f"The value is: {value}")
  else:
    print("Value not found. Please check the row and column names.")
```

```python
import ipywidgets as widgets
from IPython.display import display



# Creating dropdown widgets for row and column selection
row_options = df['Operating Results'].unique().tolist()
row_dropdown = widgets.Dropdown(
    options=row_options,
    value=row_options[0],
    description='Row:',
    disabled=False,
)

column_options = df.columns[1:].tolist()  # Exclude the 'Operating Results' column
column_dropdown = widgets.Dropdown(
    options=column_options,
    value=column_options[0],
    description='Column:',
    disabled=False,
)

# Creating an output widget to display the result
output = widgets.Output()

# Defining the function to handle the selection change event
def on_selection_change(change):
  with output:
    row_name = row_dropdown.value
    column_name = column_dropdown.value
    value = get_value(row_name, column_name)
    if value is not None:
      print(f"The value is: {value}")
    else:
      print("Value not found. Please check the row and column names.")

# Registering the event handler
row_dropdown.observe(on_selection_change, names='value')
column_dropdown.observe(on_selection_change, names='value')

# Displaying the widgets
display(row_dropdown, column_dropdown, output)
```
